# Setup

```python
vectorcounts = {}

for index, row in results_df.iterrows():
    locs = row['countries']
    for loc in locs:
        loc = loc.strip()
        if loc != "Netherlands":
            if loc not in vectorcounts:
                vectorcounts[loc] = [0]
            vectorcounts[loc][0] = vectorcounts[loc][0] + 1
```

```python
pycountry_is_dumb = pd.DataFrame(columns = ['country', 'code', 'code3' 'continent'])
pycountry_is_dumb = pycountry_is_dumb.append({'country': 'Sint Maarten', 'code': 'SX', 'code3': 'SXM', 'continent': 'NA'},ignore_index = True)

country_code_continent_df = pd.DataFrame(columns=['country', 'code', 'continent', 'count'])
for key in vectorcounts:
    count = vectorcounts[key][0]
    try:
        country_code = pc.country_name_to_country_alpha2(key, cn_name_format="default")
        country_code_3 = pc.country_name_to_country_alpha3(key, cn_name_format="default")
        continent_code = pc.country_alpha2_to_continent_code(country_code)
    except:
        if key in list(pycountry_is_dumb['country']):
            country_code = pycountry_is_dumb[pycountry_is_dumb['country'] == key]['code'][0]
            country_code_3 = pycountry_is_dumb[pycountry_is_dumb['country'] == key]['code3'][0]
            continent_code = pycountry_is_dumb[pycountry_is_dumb['country'] == key]['continent'][0]     
    country_code_continent_df = country_code_continent_df.append({'country': key, 'code': country_code, 'code3': country_code_3, 'continent': continent_code, 'count': count}, ignore_index=True)

country_code_continent_df['count'] = country_code_continent_df['count'].astype('int64')
country_code_continent_df
```

# Time Series

```python
def create_time_series(df, sample, trim):
    date_list = list(df['date'])
    one_list = []
    for i in date_list:
        one_list.append(1)
    series = pd.Series(one_list, index=date_list)
    series = series.resample(sample).sum()
    series = series[trim:]
    return series
```

## Per drug type

```python
drug_types = []
for index, row in results_df.iterrows():
    for drug_type in row['general_drugs']:
        if drug_type not in drug_types:
            drug_types.append(drug_type)
print(drug_types)

series_list = []
for drug in drug_types:
    temp_df = pd.DataFrame()
    for index, row in results_df.iterrows():
        if drug in row['general_drugs']:
            temp_df = temp_df.append(row)
    series = create_time_series(temp_df, '1D', 0)
    series_list.append(series)
    
drugs_time_series = pd.concat(series_list, axis=1)
drugs_time_series.columns = drug_types
```

```python
for drug in drugs_time_series:
        drugs_time_series[drug].resample('6M').sum().truncate(before='2000-01-31').rolling(window=6).mean().plot()
plt.title('Cases Concerning Different Drug Types Over Time', fontsize=fontsize)
plt.xlabel('Time', fontsize=fontsize)
plt.ylabel('Case count', fontsize=fontsize)
plt.legend(loc=2, prop={'size': legend_size})
plt.show()
```

```python
semi_general_drug_types = []
for index, row in results_df.iterrows():
    for drug_type in row['semi_general_drugs']:
        if drug_type not in semi_general_drug_types:
            semi_general_drug_types.append(drug_type)
print(semi_general_drug_types)

series_list = []
for drug in semi_general_drug_types:
    temp_df = pd.DataFrame()
    for index, row in results_df.iterrows():
        if drug in row['semi_general_drugs']:
            temp_df = temp_df.append(row)
    series = create_time_series(temp_df, '1D', 0)
    series_list.append(series)
    
semi_general_drugs_time_series = pd.concat(series_list, axis=1)
semi_general_drugs_time_series.columns = semi_general_drug_types
```

```python
for drug in semi_general_drugs_time_series:
        semi_general_drugs_time_series[drug].resample('6M').sum().truncate(before='2000-01-31').rolling(window=6).mean().plot()
plt.title('Cases Concerning Different Drug Types Over Time', fontsize=fontsize)
plt.xlabel('Time', fontsize=fontsize)
plt.ylabel('Case count', fontsize=fontsize)
plt.legend(loc=2, prop={'size': legend_size})
plt.show()
```

## Per country || per continent

```python
country_code_continent_df
```

```python
country_list = list(country_code_continent_df['country'])

series_list = []
for country in country_list:
    temp_df = pd.DataFrame()
    for index, row in results_df.iterrows():
        if country in row['countries']:
            temp_df = temp_df.append(row)
    series = create_time_series(temp_df, '1D', 0)
    series_list.append(series)
  

country_time_series = pd.concat(series_list, axis=1)
country_time_series.columns = country_list
```

Get x countries with most cases

```python
count = 10
country_counts = country_time_series.sum().sort_values()[-count:]
print(country_counts)
biggest_countries = list(country_counts.index)[::-1]
```

```python
for country in country_time_series[biggest_countries]:
        country_time_series[country].resample('6M').sum().truncate(before='2000-01-31').rolling(window=6).mean().plot()
plt.title('Cases Concerning Different Countries Over Time', fontsize=fontsize)
plt.xlabel('Time', fontsize=fontsize)
plt.ylabel('Case count', fontsize=fontsize)
plt.legend(loc=2, prop={'size': legend_size})
plt.show()
```

```python
continent_list = []
for continent in country_code_continent_df['continent']:
    if continent not in continent_list:
        continent_list.append(continent)
 
continent_time_series = pd.DataFrame(columns=continent_list, index = country_time_series.index)

for country in country_time_series:
    continent = list(country_code_continent_df[country_code_continent_df['country'] == country]['continent'])[0]
    
    list_a = [0 if math.isnan(x) else x for x in list(country_time_series[country])]
    list_b = [0 if math.isnan(x) else x for x in list(continent_time_series[continent])]
    continent_time_series[continent] = [a+b for a, b in zip(list_a, list_b)]
               
          
```

```python
for continent in continent_time_series:
        continent_time_series[continent].resample('6M').sum().truncate(before='2000-01-31').rolling(window=6).mean().plot()
plt.title('Cases Concerning Different Continents Over Time', fontsize=fontsize)
plt.xlabel('Time', fontsize=fontsize)
plt.ylabel('Case count', fontsize=fontsize)
plt.legend()
plt.show()
```
